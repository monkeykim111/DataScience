import를 이용하여 외부 라이브러리 사용하기 & 에너지 데이터 불러오기


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objects as go

df = pd.read_excel('/content/energy.xlsx')
df.head()
```

```python
df.tail()
```

```python
df.dtypes
```

테이블의 첫째 열을 'day'로 보고 싶다면</br>
set_index('컬럼 명') 을 하면 해당 컬럼 명이 인덱스가 된다.

```python
df.set_index('day')
```

전체 테이블에서 원하는 열만 가져와서 또 다른 파일 만들기

```python
df1 = df[['day', '환기', '콘센트', '냉방', '난방', '조명', '기타']]
df1.head()
```

```python
df1.to_excel('newfile.xlsx')
# df1의 데이터가 엑셀로 저장이 됨!! - 다운로드
```

2. 값 비교하기

```python
df[df.month==10]
```

```python
df[df.환기 >= 2]
```

3. and, or를 이용하여 여러 조건으로 파일 불러오기

```python
df2 = df[(df.month == 10) & (df.weekday == 'Saturday')]
df2
```

자기 생일이 있는 월을 선택하여 냉방이 1보다 크거나 같으면 df4라는 변수에 저장하고, 저장한 파일을 불러오시오.

```python
df3 = df[(df.month == 12) & (df.냉방 >= 1)]
df3.head()
```

```python
df3.to_excel('12월냉방.xlsx')
```

데이터 다시 불러오기

```python
df4 = pd.read_excel('/content/12월냉방.xlsx')
df4
```

일별 환기 산점도 그래프 시각화

```python
fig = go.Figure(go.Scatter(
    x = df['day'],
    y = df['환기'],
))
fig.show()
```

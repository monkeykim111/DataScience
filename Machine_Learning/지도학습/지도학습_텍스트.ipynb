###분류와 회귀
* 지도 학습에는 분류와 회귀가 있음</br>

**분류**: 미리 정의된, 가능성이 있는 여러 클래스 레이블 중 하나를 예측하는 것</br>

*이진분류:* 딱 두 개의 클래스로 분류</br>
 - 이진분류에서 한 클래스를 양성 클래스, 다른 하나를 음성 클래스로 분류</br>
(양성클래스: 좋은 값이나 장점을 나타내는 것이 아니라 학습하고자 하는 대상을 말함 // ***내가 찾고자 하는 값을 1로 표시한다!!***)</br>

*다중분류:* 셋 이상의 클래스로 분류</br>
 - 붓꽃의 분류, 웹사이트의 글로부터 어떤 언어의 웹사이트인지를 예측(한국어, 영어, 일본어 ...)



**회귀**: 연속적인 숫자, 또는 프로그래밍 용어로 부동소수점수(실수)를 예측하는 것
- 어떤 사람의 교육수준, 나이, 주거지를 바탕으로 연간 소득을 예측
- 옥수수 농장에서 전년도 수확량, 날씨, 고용 인원수 등으로 올해 수확량을 예측

**출력값에 연속성**이 있는지 질문해보면 회귀와 분류 문제를 쉽게 구분할 수 있음
- 출력값 사이에 연속성이 있다면 회귀문제, 연속성이 없다면 분류문제임

### 일반화 성능이 최대가 되는 모델이 최적의 모델이다.
* 지도 학습에서는 훈련 데이터로 학습한 모델이 훈련 데이터와 특성이 같다면 처음 보는 새로운 데이터가 주어져도 정확히 예측할 것이라 기대함

* 모델이 처음보는 데이터에 대해 정확하게 예측할 수 있음 -> 훈련 세트에서 테스트 세트로 일반화 되었다고 함

* 보통 훈련 세트에 대해 정확히 예측하도록 모델을 구축함 -> 훈련 세트와 테스트 세트가 매우 비슷하다면 그 모델이 테스트 세트에서도 정확히 예측하리라 기대할 수 있음

* 가진 정보를 모두 사용해서 너무 복잡한 모델을 만드는 것은 과대적합, 너무 간단한 모델이 선택되는 것은 과소적합이다.


## 일반화, 과대적합, 과소적합
모델 복잡도와 데이터셋 크기의 관계

모델을 복잡하게 할수록 훈련 데이터에 대해서는 더 정확하게 예측이 가능함</br>
하지만 너무 복잡해지면 훈련 세트의 각 데이터 포인트에 너무 민감해져 새로운 데이터에 일반화되기 어려움</br>
-> 과대적합

일반화 성능이 최대가 되는 최적점에 있는 모델이 우리가 찾으려는 모델임

모델의 복잡도는 훈련 데이터셋에 담긴 입력 데이터의 다양성과 관련이 있음(다양한 피쳐들을 가지는 게 중요)</br>
-> 데이터셋에 다양한 데이터 포인트가 많을수록 과대적합이 없어 더 복잡한 모델을 만들 수 있음

***같은 데이터 포인트를 중복하거나 매우 비슷한 데이터를 모으는 것은 도움이 되지 않는다.***</br>

-> ***데이터를 더 많이 수집하고 적절하게 더 복잡한 모델을 만들면 지도학습 문제에서 종종 놀라운 결과를 얻을 수가 있다.***

문제정의: 도매업 고객(Wholesale customers) 데이터 셋을 활용한 군집 분석

```python
# 기본 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt
```

step01. 데이터 준비하기

```python
# Wholesale customers 데이터셋 가져오기 (출처: UCI ML Repository)
uci_path = 'https://archive.ics.uci.edu/ml/machine-learning-databases/\
00292/Wholesale%20customers%20data.csv'

df = pd.read_csv(uci_path, header= 0)
```

step02. 데이터 탐색하기

```python
# 데이터 살펴보기 (연간 데이터 셋임)
df.head() 
```

```python
# 데이터 자료형 확인
df.info()
```

step03 데이터 전처리하기

```python
# 데이터가 너무 크므로 데이터 전처리가 필요함

# 군집 분석은 모든 속성이 X (정답 y는 없다.)

X = df.iloc[:, :] # 전체 데이터를 사용하겠다.
print(X[:5])
```

```python
# 데이터 정규화하기

from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
print(X[:5])
```

step04. k-means 군집 모형 - sklearn 사용


```python
# sklearn 라이브러리에서 cluster 군집 모형 가져오기
from sklearn import cluster

# https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html
# 모형 객체 생성 클러스터 개수 5개 지정

kmeans = cluster.KMeans(init='k-means++', n_clusters=5, n_init=10)
```

```python
# 모형 학습하기

kmeans.fit(X)
```

```python
# 예측 (군집)
cluster_label = kmeans.labels_ # 0 ~ 4까지 군집
print(cluster_label)
```

```python
# 예측 결과를 데이터 프레임에 추가하기

df['Cluster'] = cluster_label
df.head()
```

클러스터 데이터 시각화

```python
# 그래프로 표현 - 시각화
df.plot(kind='scatter', x='Grocery', y='Frozen', c='Cluster', cmap='Set1', figsize=(10, 10))

plt.show()
```

```python
# 그래프로 표현 - 시각화
df.plot(kind='scatter', x='Fresh', y='Milk', c='Cluster', cmap='Set1', figsize=(10, 10))

plt.show()
```

```python
# 큰 값으로 구성된 클러스터(0, 4)를 제외 - 값이 몰려 있는 구간을 자세하게 분석

mask = (df['Cluster'] == 0) | (df['Cluster'] == 4)
ndf = df[~mask] # '~' == not을 의미
ndf.plot(kind='scatter', x='Grocery', y='Frozen', c='Cluster', cmap='Set1', figsize=(10, 10))
plt.show()
```

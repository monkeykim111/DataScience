```python
# 라이브러리 import
import pandas as pd
import tensorflow as tf
```

```python
# data 불러오기
Iris = pd.read_csv("/content/iris.csv")
Iris.head()
```

```python
# 원핫 인코딩
encoded_Iris = pd.get_dummies(Iris)
encoded_Iris
```

```python
encoded_Iris.info()
```

```python
encoded_Iris.columns
```

```python
# 독립변수, 종속변수 분리하기

독립 = encoded_Iris[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
종속 = encoded_Iris[['품종_setosa', '품종_versicolor', '품종_virginica']]
```

```python
print(독립.shape, 종속.shape)
```

```python
# 모델 생성
X = tf.keras.layers.Input(shape=[4])
H = tf.keras.layers.Dense(5, activation='swish')(X)
H = tf.keras.layers.Dense(5, activation='swish')(H)
H = tf.keras.layers.Dense(5, activation='swish')(H)
y = tf.keras.layers.Dense(3, activation='softmax')(H)

model = tf.keras.models.Model(X, y)
model.compile(loss="categorical_crossentropy",
              metrics='accuracy')
```

```python
model.summary()
```

```python
# 모델 학습
model.fit(독립, 종속, epochs=600, verbose=0)
model.fit(독립, 종속, epochs=10)
# 학습이 되고있는지를 확인해 보는 척도는 accuracy보단 loss를 확인할 것
```

```python
# 모델 예측

model.predict(독립[:10])
# 5번째 인덱스까지 확인했을 때, 
# 첫번째 컬럼인 품종_setosa를 9.8이상으로 예측, 두번째, 3번째 컬럼은 1에 가깝게 예측을 함 
```
